## Deep Neural Network in Keras

```python
import numpy as np
np.random.seed(42)
```

```python
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers.normalization import BatchNormalization
from keras import regularizers
from keras.optimizers import SGD
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
X_train = X_train.reshape(60000,784).astype('float32')
X_test = X_test.reshape(10000, 784).astype('float32')
```

```python
X_train /= 255
X_test /= 255
```

```python
n_classes = 10
y_train = keras.utils.to_categorical(y_train, n_classes)
y_test = keras.utils.to_categorical(y_test, n_classes)
```

```python
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(784,),
                kernel_initializer='glorot_normal', kernel_regularizer=regularizers.l2(0.01)))
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer=SGD(lr=0.1), metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, batch_size=128, epochs=100, verbose=1, validation_data=(X_test, y_test))
```

## Dropout 
To run the model restart the kernel

```python
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(784,),)) # we use default keras initilization 
model.add(Dropout(0.5)) # Dropout is ised between layers
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer=SGD(lr=0.1), metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, batch_size=128, epochs=100, verbose=1, validation_data=(X_test, y_test))
```

```python
## No dropput during validation, allneuros will be used for validation
## There reason wht trainig and test set accuracy have some accuracy difference
```

## droput + Regularization

```python
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(784,), kernel_regularizer=regularizers.l2(0.01))) # we use default keras initilization 
model.add(Dropout(0.5)) # Dropout is ised between layers
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer=SGD(lr=0.1), metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, batch_size=128, epochs=100, verbose=1, validation_data=(X_test, y_test))
```

```python
# When we use dropout + regularization validation accuracy dropped a little bit
# Based on experience just dropout is enough, no need to use together
```

## BatchNormalization Layer

```python
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(784,),)) # we use default keras initilization 
model.add(BatchNormalization())
model.add(Dropout(0.5)) # Dropout is ised between layers
model.add(Dense(64, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer=SGD(lr=0.1), metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, batch_size=128, epochs=100, verbose=1, validation_data=(X_test, y_test))
```

## Second-order calculation (Adam optimizer)
